Build a quadratic function $f(x)=0.1(x*x) - 9x + 4500$ on $x ∈ [0, 10]$to create the per-unit cost function, and make a conclusion.


##Exercise 2:
Sum of two lists

There will be many situations in which we'll have to find an index-wise summation of two different lists. This can have possible applications in day-to-day programming. In this exercise, we will solve the same problem in various ways in which this task can be performed.

We have the following two lists:

list1 = [2, 5, 4, 7, 3]
list2 = [1, 4, 6, 9, 10]

Now let's use Python code to demonstrate addition of two lists.


```python
list1 = [2, 5, 4, 7, 3]
list2 = [1, 4, 6, 9, 10]
```

```python
# metodo 1 usando un bucle for y zip
suma = []
for a, b in zip(list1, list2):
    suma.append(a + b)
print ("list 1 : " + str(list1))
print ("list 2 : " + str(list2))
print("Suma de las listas usando bucle for:", suma)
```

```python
# metodo 2 usando list cromprehension
suma = [a + b for a, b in zip(list1, list2)]
print ("list 1 : " + str(list1))
print ("list 2 : " + str(list2))
print("Suma usando list comprehension:", suma)
```

```python
# usando numpy mas rapido con listas grandes
suma = np.array(list1) + np.array(list2)
print ("list 1 : " + str(list1))
print ("list 2 : " + str(list2))
print("Suma usando NumPy:", suma.tolist())
```


##Exercise 3:
Dot multiplication

We have two matrices:

matrix1 = [[1,7,3],
 [4,5,2],
 [3,6,1]]
matrix2 = [[5,4,1],
 [1,2,3],
 [4,5,2]]

A simple technique but expensive method for larger input datasets is using for loops. In this exercise, we will first use nested for loops to iterate through each row and column of the matrices, and then we will perform the same multiplication using NumPy.


```python
matrix1 = [[1, 7, 3],
           [4, 5, 2],
           [3, 6, 1]]

matrix2 = [[5, 4, 1],
           [1, 2, 3],
           [4, 5, 2]]
resultado = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]

for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
        for k in range(len(matrix2)):
            resultado[i][j] += matrix1[i][k] * matrix2[k][j]

print("resultado con for loops:")
for fila in resultado:
    print(fila)          
```

```python
# metodo con numpy
matrix1 = [[1, 7, 3],
           [4, 5, 2],
           [3, 6, 1]]

matrix2 = [[5, 4, 1],
           [1, 2, 3],
           [4, 5, 2]]
resultado_np = np.dot(matrix1, matrix2)
print("resultado con numpy:")
print(resultado_np)
```
